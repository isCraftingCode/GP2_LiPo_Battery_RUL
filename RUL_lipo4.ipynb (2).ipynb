```python
import pandas as pd
```

```python
# قراءة الملف
df = pd.read_excel("Capacity_stress_outputكامله العداد.xlsx")
```

```python
# التأكد من أن البيانات رقمية

df["Capacity"] = pd.to_numeric(df["Capacity"], errors="coerce")
df["Cycle"] = pd.to_numeric(df["Cycle"], errors="coerce")
```

```python
# حذف أي صفوف فيها بيانات ناقصة
df.dropna(subset=["Capacity", "Cycle"], inplace=True)
```

```python
# حساب السعة الابتدائية
initial_capacity = df["Capacity"].iloc[0]
```

```python

# تحديد العتبة لنهاية العمر (80% من السعة الابتدائية)
eol_threshold = 0.8 * initial_capacity
```

```python
# إيجاد أول دورة أقل من العتبة
eol_cycle = df[df["Capacity"] < eol_threshold].iloc[0]["Cycle"]
```

```python
# حساب RUL لكل دورة
df["RUL"] = eol_cycle - df["Cycle"]

```
