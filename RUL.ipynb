```python
import pandas as pd
```

```python
capacity_df = pd.read_csv("Capacity_std.csv", sep="\t", header=None)
capacity_df.columns = ["Cycle", "Capacity"]
```

```python
capacity_df.head()
```

```python
initial_capacity = capacity_df["Capacity"].iloc[0]
eol_threshold = 0.8 * initial_capacity
eol_cycle = capacity_df[capacity_df["Capacity"] < eol_threshold].iloc[0]["Cycle"]

# Compute RUL per cycle
capacity_df["RUL"] = eol_cycle - capacity_df["Cycle"]
```

```python
capacity_df.head()
```

```python
df = pd.read_csv("0_0_1_stddischarge.csv", sep="\t", header=None)
```

```python
df.columns = ["Time (s)", "Current (A)", "Voltage (V)"]
```

```python
df.head()
```

```python

T = df["Time (s)"].max()
RUL_start = capacity_df.loc[capacity_df["Cycle"] == 0, "RUL"].values[0]
df["RUL"] = RUL_start - (df["Time (s)"] / T)
```
