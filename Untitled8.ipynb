```python
import pandas as pd
```

```python
max_capacity = 1.1
drop_rate = 0.00107843
```

```python
cycles = list(range(0, 205))
capacities = []
rul_values = []
for n in cycles:
  x = max_capacity - (n * drop_rate)
  rul = (x/max_capacity) * 100
  capacities.append(x)
  rul_values.append(rul)

df = pd.DataFrame({
    "Cycle": cycles,
    "Estimated Capacity": capacities,
    "RUL (%)": rul_values})

df
```

```python
df.to_excel("estimated_rul_cycles_to204.xlsx", index=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(df["Cycle"], df["RUL (%)"], marker='o', linestyle='-', color='blue')

plt.title("Remaining Useful Life (RUL) per Cycle")
plt.xlabel("Cycle Number")
plt.ylabel("RUL (%)")
plt.grid(True)
plt.xticks(range(0, 46, 1))
plt.ylim(0, 110)


plt.tight_layout()
plt.show()
```
